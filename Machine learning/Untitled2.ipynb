```python

import json
import requests


def pm10(city, location):
	location = location.strip(' ')
	b = ''
	for i in location:
		if i == ' ':
			b+='%20'
		else:
			b+=i
	url = 'https://api.openaq.org/v1/measurements?city='+city+'&location='+b+'&parameter=pm10'
	result = requests.get(url).content
	print result
	result = json.loads(result)
	if result['meta']['found'] == 0:
		return 'No data found for this location.'
	else:
		return result['results'][0]['value']

def generate_response(msg):
	msg= msg.split(',')
	location=msg[0]
	city=msg[1]
	result = pm10(city,location)
	if result != 'No data found for this location.':
		print 'Hi, The pm10 value for this location is '+str(result)
	else:
		print result

generate_response('Anand Vihar,Delhi')
```
